```python
#common code for reading
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/cllg assignment/AssignNo4comp_sales_data.csv')
df
```

**Total profit For Each Month**

```python
x = df['month_number'].tolist()
y = df['total_profit'].tolist()
plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.title('total profit for each month')
plt.plot(x,y)
plt.show
```

**Total profit For Each Month in a different style**

```python
x = df['month_number'].tolist()
y = df['total_units'].tolist()
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.title('total profit for each month')
plt.plot(x,y,label='Sold units number',linestyle = 'dotted',linewidth = '3',color = 'red', marker = 'o', )
plt.legend(loc='lower right')
plt.show
```

**All product sales data on a multiline plot**

```python
x = df['month_number'].tolist()
y1 = df['facecream'].tolist()
y2 = df['facewash'].tolist()
y3 = df['toothpaste'].tolist()
y4 = df['bathingsoap'].tolist()
y5 = df['shampoo'].tolist()
y6 = df['moisturizer'].tolist()
plt.xlabel('Month Number')
plt.ylabel('no of units')
plt.plot(x,y1,label='facecream')
plt.plot(x,y2,label='facewash')
plt.plot(x,y3,label='toothpaste')
plt.plot(x,y4,label='bathingsoap')
plt.plot(x,y5,label='shampoo')
plt.plot(x,y6,label='moisturizer')
plt.legend(loc='upper left')
plt.title('Number of units sold per month')
plt.show
```

**Toothpaste sales data of each month and shown using a scatter plot**

```python
x = df['month_number'].tolist()
y = df['toothpaste'].tolist()
plt.xlabel('Month Number')
plt.ylabel('Toothpaste sold')
plt.scatter(x,y)
plt.grid(color = 'black', linestyle = '-', linewidth = 0.5)
plt.title('toothpaste sales data of each months')
plt.show
```

**face cream and facewash product sales data and shown using the bar chart**

```python
import numpy as np
x = df['month_number'].tolist()
y1 = df['facecream'].tolist()
y2 = df['facewash'].tolist()

X_axis = np.arange(len(x))
  
plt.bar(X_axis - 0.2, y1, 0.4, label = 'facecream')
plt.bar(X_axis + 0.2, y2, 0.4, label = 'facewash')
  
plt.xticks(X_axis, x)
plt.xlabel("number of month")
plt.ylabel("sales")
plt.title("sales of facecream and facewash")
plt.legend()
plt.show()
```

**Sales data of bathing soap of all months**

```python
x = df['month_number'].tolist()
y = df['bathingsoap'].tolist()
plt.xlabel('Month Number')
plt.ylabel('bathing soap')
plt.title('sales of bathing soap')
plt.bar(x,y)
plt.show
```

**Total sale data for last year for each product shown using a Pie chart**

```python
total = df[['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']].sum()
df
df.loc[12] = total
df
grp = df.values.tolist()
grp
total = grp[12]
sales = [total[1],total[2],total[3],total[4],total[5],total[6]]
products = ['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
plt.pie(sales , labels = products , autopct ='%1.1f%%',startangle = 90)
plt.title('Total sales data')
plt.show()
```
